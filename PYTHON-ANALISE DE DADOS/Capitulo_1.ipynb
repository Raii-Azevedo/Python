# 📘 Python para Análise de Dados — Capítulo 1  
_📖 Por Wes McKinney_

Há tempos venho querendo me aprofundar em análise de dados utilizando exclusivamente os recursos do ecossistema Python. Para isso, escolhi o clássico _“Python for Data Analysis”_, de Wes McKinney, uma referência obrigatória para quem trabalha com ciência de dados, modelagem e análise.

## 🧠 Capítulo 1 — Uma Introdução Valiosa

Apesar de introdutório, o primeiro capítulo vai além das já conhecidas origens da linguagem. McKinney apresenta uma visão estratégica sobre **as vantagens e limitações do Python para análise de dados**, além de compará-lo com outras linguagens populares no campo:

### 🔍 Destaques do capítulo:

- Comparativo com linguagens concorrentes como **R, MATLAB e SAS**;
- A principal vantagem do Python: sua capacidade de integração com linguagens de baixo nível (**C, C++ e Fortran**), o que o torna ideal para lidar com **código legado**;
- Python resolve bem o problema das “duas linguagens” (uma para prototipar, outra para implementar), permitindo tanto **experimentação rápida quanto desenvolvimento robusto** em produção;
- Embora o autor recomende o uso do **IPython**, optarei pelo **Jupyter Notebook**, por sua ótima integração com **Markdown** e capacidade de gerar **documentação interativa e bem estruturada**.

## 🧰 Pacotes e Bibliotecas

Também são apresentados os principais **pacotes e bibliotecas** que compõem o stack de ciência de dados em Python:

- `NumPy`, `Pandas`, `Matplotlib`, `SciPy` — fundamentais para computação científica e manipulação de dados;
- O que mais me chamou atenção: a introdução ao **Scikit-learn**, biblioteca voltada para _machine learning_, com aplicações práticas como:
  - Regressão
  - Clustering
  - Seleção de modelos

> ⚠️ Embora o livro não aprofunde tanto no Scikit-learn, a menção já é um ótimo ponto de partida para explorar mais.
